```python
import pandas as pd
import numpy as np

path = './data/slam_data.csv'
df = pd.read_csv(path)
df.fillna(0, inplace=True)
df.set_index('img_id', drop=True, inplace=True)
df.drop(columns=['Unnamed: 0'], inplace=True) # drop the index from the dataframe that wrote the csv data
```

```python
import matplotlib.pyplot as plt

plt.figure()
df.plot()
plt.ylabel('Count')
plt.legend(loc='best')
```
